```python
import os

```

```python
def get_doc_text(file):
    if file.endswith('.doc') or file.endswith('.docx'):
       # converting .doc to .docx
        doc_file = file
        print(doc_file)
        if file.endswith('.doc'):
            new_file = doc_file.replace(".doc",".txt")
        if file.endswith('.docx'):
            new_file = doc_file.replace(".docx",".txt")
        #this line make linux can read space
        doc_file = doc_file.replace(" ", "\ ")
        new_file = new_file.replace(" ","_")
        print(new_file)
        os.system('antiword -f ' + doc_file  + ' > ' + new_file)
        os.remove(file)
#         with open(new_file) as f:
#               text = f.read()
#         print(text)

#         input("")
```

```python

def prune_text(file):
    if(' ' in file):
        sourceFileName = file.replace(" ","\ ")
        targetFileName = file.replace(" ","_")
        os.system('iconv -f BIG-5 -t UTF-8 '+ sourceFileName + ' > ' + targetFileName)
        os.remove(file)
        print("%s is deleted, %s is created"%(sourceFileName,targetFileName))
        
```

```python
for root, dirs, files in os.walk("."):
    for file in files:
        path = os.path.join(root, file)
        get_doc_text(path)
        prune_text(path)
```
